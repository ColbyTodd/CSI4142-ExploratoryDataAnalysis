# German Credit Dataset
## Description
The original dataset contains 1000 entries with 20 categorial/symbolic attributes prepared by Prof. Hofmann. In this dataset, each entry represents a person who takes a credit by a bank. Each person is classified as good or bad credit risks according to the set of attributes. The link to the original dataset can be found below.
 - Author: Prof. Hofmann
 - Purpose: Determing credit risk of a person
 - Shape: 10 columns, 1000 rows
 - Description: Financial data for credit scoring
 - Features:
   - Age (Numerical)
   - Sex (Categorical)
   - Job (Categorical)
   - Housing (Categorical)
   - Saving Accounts (Categorical)
   - Checking Account (Numerical)
   - Credit Amount (Numerical)
   - Duration (Numerical)
   - Purpose (Categorical)

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import kagglehub
from kagglehub import KaggleDatasetAdapter

# Load German Credit Dataset into a pandas dataframe
credit_df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "uciml/german-credit/versions/1",
    "german_credit_data.csv",
)

credit_df.info()
credit_df.head
```

## Insight 1
Most people have less than DM5000 of credit.

```python
sns.histplot(credit_df['Credit amount'])
```

A univariate analysis was done using a histogram (r1).

This evidence was obtained by looking at the histogram.

## Insight 2
There is over double the amount of men than women.

```python
sns.countplot(x="Sex", data=credit_df)
```

This evidence was obtained by looking at the countplot.

## Insight 3
There is no correlation between crediamount and age.

```python
sns.scatterplot(x='Credit amount', y='Age', data=credit_df)
```

A bivariate analysis using a scatterplot (r6).

This evidence was obtained by looking at the scatterplot.

## Insight 4
The majority of people are skilled workers who own their home.

```python
sns.countplot(x="Job", hue="Housing", data=credit_df, 
                  order=credit_df[credit_df['Housing'] == 'own']
                    .groupby('Job')
                    .size()
                    .sort_values(ascending=False)
                    .index
             )
```

A bivariate analysis using an ordered countplot (r5).

This evidence was obtained by looking at the countplot.
